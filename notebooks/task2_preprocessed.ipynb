# Предобработка данных (task 2)

```python
import pandas as pd
```

1. Загрузка данных в DataFrame

```python
students_df = pd.read_csv('students.csv', sep=";")
students_df.head(10)
```

2. Заполнение пропусков

```python
# информация о данных
students_df.info()
```

```python
# сумма нулевых значений по столбцам (все необходимые данные корректно заполнены)
students_df.isnull().sum()
```

```python
# заполение нулевых значений средними значениями
students_df.fillna(students_df.mean())
```

3. Перекодирование категориальных признаков

```python
# работа матери (категориальный признак: "teacher", "health", "services", "at_home", "other")
pd.get_dummies(students_df, prefix="mj", columns=["Mjob"]).iloc[:10, -5:]
```

```python
# работа отца (категориальный признак: "teacher", "health", "services", "at_home", "other")
pd.get_dummies(students_df, prefix="fj", columns=["Fjob"]).iloc[:10, -5:]
```

```python
# причина выбора данной школы (категориальный признак: "home", "reputation", "course", "other");
pd.get_dummies(students_df, prefix="r", columns=["reason"]).iloc[:10, -4:]
```

```python
# опекун ученика
pd.get_dummies(students_df, prefix="g", columns=["guardian"]).iloc[:10, -3:]
```

4. Перекодирование порядковых признаков (все порядковые признаки являются упорядоченными множествами неотрицательных натуральных чисел)

5. Приведение бинарных признаков к 1 или 0

```python
# школа
students_df["school"] = students_df["school"].map({
    "GP": 0,
    "MS": 1
})
students_df["school"]
```

```python
# пол ученика
students_df["sex"] = students_df["sex"].map({
    "M": 0,
    "F": 1
})
students_df["sex"]
```

```python
# тип домашнего адреса ученика
students_df["address"] = students_df["address"].map({
    "U": 0, # 'U' - городской
    "R": 1  # 'R' - сельский
})
students_df["address"]
```

```python
# количество человек в семье
students_df["famsize"] = students_df["famsize"].map({
    "LE3": 0, # 'LE3' - меньше или равно 3
    "GT3": 1  # 'GT3' - больше 3
})
students_df["famsize"]
```

```python
# статус совместного проживания родителей
students_df["Pstatus"] = students_df["Pstatus"].map({
    "T": 0, # 'T' - совместное проживание
    "A": 1  # 'A' - порознь
})
students_df["Pstatus"]
```

```python
# занимается ли ученик дополнительным образованием
students_df["schoolsup"] = students_df["schoolsup"].map({
    "no": 0,
    "yes": 1
})
students_df["schoolsup"]
```

```python
# помогает ли семья в учебе
students_df["famsup"] = students_df["famsup"].map({
    "no": 0,
    "yes": 1
})
students_df["famsup"]
```

```python
# берет ли ученик дополнительные платные занятия в рамках предмета курса
students_df["paid"] = students_df["paid"].map({
    "no": 0,
    "yes": 1
})
students_df["paid"]
```

```python
# участвует ли ученик во внеклассных школьных мероприятиях
students_df["activities"] = students_df["activities"].map({
    "no": 0,
    "yes": 1
})
students_df["activities"]
```

```python
# посещал ли ученик детский сад
students_df["nursery"] = students_df["nursery"].map({
    "no": 0,
    "yes": 1
})
students_df["nursery"]
```

```python
# планирует ли ученик получать высшее образование
students_df["higher"] = students_df["higher"].map({
    "no": 0,
    "yes": 1
})
students_df["higher"]
```

```python
# имеет ли ученик доступ в интернет
students_df["internet"] = students_df["internet"].map({
    "no": 0,
    "yes": 1
})
students_df["internet"]
```

```python
# состоит ли ученик в романтических отношениях
students_df["romantic"] = students_df["romantic"].map({
    "no": 0,
    "yes": 1
})
students_df["romantic"]
```

6. Сохранение отредактированного датасета

```python
students_df.to_csv('students_preprocessed.csv')
```
